```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
import csv

url = 'https://wiki.mcs.anl.gov/leyffer/index.php/MacMPEC'
html = urllib2.urlopen(url).read()
soup = BeautifulSoup(html)
tables = soup.find_all('table')

table = tables[1]

# python3 just use th.text
headers = [th.text for th in table.select("tr th")]


with open("out.csv", "w") as f:
    row = ""
    i = 0
    for header in headers:
        if (i > 0):
            row += ","
        print(row)
        row += header.strip()
        i += 1
            
    f.write(row + "\n")
    
    for row_elements in table.select("tr"):
        row = ""
        i = 0    
        for td in row_elements.find_all("td"):
            if (i > 0):
                row += ","
            
            row += td.text.strip()
            i += 1
            
        print(row)
        f.write(row + "\n")
```
